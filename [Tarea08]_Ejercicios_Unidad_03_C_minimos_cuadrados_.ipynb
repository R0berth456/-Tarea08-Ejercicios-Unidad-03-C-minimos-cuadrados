Primer Literal

a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(f'm = {alpha[0]}, b = {alpha[1]}')

#Calcular el error
error = 0
for i in range(len(x)):
    error += (y[i] - alpha[0]*x[i] - alpha[1])**2
print(f'El error es {error}')

plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

b. Construya el polinomio por mínimos cuadrados de grado 2 y calcule el error.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('seaborn-poster')

def func(x, a, b, c):
    y = a*x**2 + b*x + c
    return y

x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

alpha1 = curve_fit(func, x, y)[0]
print(f'a = {alpha1[0]}, b = {alpha1[1]}, c = {alpha1[2]}')

#Calcular el error
error = 0
for i in range(len(x)):
    error += (y[i] - alpha1[0]*x[i]**2 - alpha1[1]*x[i] - alpha1[2])**2
print(f'El error es {error}')

plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha1[0]*x**2 + alpha1[1]*x + alpha1[2], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
